# Pre-calculate Embeddings

```python
from sentence_transformers import SentenceTransformer

## Precalculate Embedding
embedding_model = SentenceTransformer("multi-qa-MiniLM-L6-dot-v1")
```

```python
print(preprocessed_df.columns)
```

```python
preprocessed_df.columns = preprocessed_df.columns.str.strip()
```

```python
preprocessed_df['Text'] = preprocessed_df['Text'].str.strip()

# Replace multiple spaces between words with a single space
preprocessed_df['Text'] = preprocessed_df['Text'].str.replace(r'\s+', ' ', regex=True)
```

```python
# Try encoding
try:
    embeddings_Text = embedding_model.encode(preprocessed_df['Text'], show_progress_bar=True)
except KeyError as e:
    print(f"KeyError encountered: {e}")
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
from umap import UMAP
umap_model = UMAP(n_neighbors=10, n_components=5, min_dist=0.1, metric='euclidean')

from hdbscan import HDBSCAN
hdbscan_model = HDBSCAN(min_cluster_size=100, min_samples = 120, metric='euclidean', cluster_selection_method='eom', prediction_data=True)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer_model = CountVectorizer(stop_words="english", min_df=2, ngram_range=(1, 2))
```

```python
from bertopic import BERTopic
topic_model = BERTopic(
    embedding_model = embedding_model,
    umap_model = umap_model,
    hdbscan_model = hdbscan_model,
    vectorizer_model = vectorizer_model,
    nr_topics = 13,
    top_n_words = 20,
    verbose = True
)
```

```python
## Topics Generation
topics, probs = topic_model.fit_transform(preprocessed_df['Text'], embeddings_Text)
topic_model.get_topic_info()
```

```python
topic_model.get_topic_info().to_csv('../../Replication Package/Data/Output/topics_info-QA-Model_GitHub_Issues_PullRequests.csv')
```

# Document Distribution

```python
document_distribution = topic_model.get_document_info(preprocessed_df['Text'])
document_distribution
```

```python
concatenated_df = pd.concat([preprocessed_df, document_distribution], axis=1)
concatenated_df
```

```python
distinct_topics = concatenated_df['Topic'].unique()
distinct_topics
```

```python
for topic in distinct_topics:
    df_topic = concatenated_df[concatenated_df['Topic'] == topic]
    filename = f"{topic}_Topics_data.csv"
    df_topic.to_csv('../../Replication Package/Data/Output/Topic Distribution GitHub/' + filename)
```
